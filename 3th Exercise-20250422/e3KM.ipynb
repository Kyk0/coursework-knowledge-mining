#### Cross-validation

a) To perform 10-fold cross-validation means to divide the dataset on 10 equal parts, train the model on 9 of them, and test it on the one part that is left. After that, repeat the process 9 more times with all the parts. In the end, the results are averaged to check how good the model performs.

b) Cross-validation is used to check how good a model performs on new data and helps detect overfitting or underfitting.
