#### Part 2: Association rules

a) **Association rules** are patterns that show which items usually appear together and are used to find relationships in the data.


b) **Support** is the proportion of transactions which contains specified items.

**Confidence** is the probability that one item appears in a transaction when another item is already appeared.


c) **The Apriori algorithm** is used to find frequent itemsets and generate association rules. It first finds items that appear often and combines them into bigger groups if they meet the support threshold. Overall, the idea is that all parts of a frequent itemset has to be frequent as well.
